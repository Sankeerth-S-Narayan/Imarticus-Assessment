#### 1. There are '234' different countries present in this data

```python
a=df.groupby('Continent')['Area (km²)'].sum()
a=pd.DataFrame(data=a)
a=a.reset_index()
```

```python
a
```

```python
sns.catplot(x="Continent", y="Area (km²)",kind="bar", data=a,aspect=2)
```

#### 2. Asia has the maximum area in sq km

```python
b=df.groupby('Country')['Growth Rate'].sum()
b=pd.DataFrame(data=b)
b=b.reset_index()
```

```python
c=b.sort_values(by=['Growth Rate'],ascending=False).head(5)
```

```python
c
```

```python
ax = plt.axes()
sns.barplot(x="Growth Rate", y="Country", data=c,ci = 95,orient='h')
ax.set(ylabel=None)
```

#### 3.Moldova has the highest growth rate of '1.0691' 

```python
d=df.groupby('Country')['World Population Percentage'].sum()
d=pd.DataFrame(data=d)
d=d.reset_index()
e=d.sort_values(by=['World Population Percentage'],ascending=False).head(6)
f=df.groupby('Continent')['World Population Percentage'].sum()
f=pd.DataFrame(data=f)
f=f.reset_index()
```

```python
plot1 = plt.figure(1)
plt.bar(e['Country'],e['World Population Percentage'])
plot2 = plt.figure(2)
plt.bar(f['Continent'],f['World Population Percentage'])
plt.show()
```

#### 4.Asia has the maximum share in the World population among continents whilst India and China occupy the top positions with highest population percentage among countries

```python
g=df[['Country','2022 Population','1970 Population']]
```

```python
g=g.sort_values(by=['2022 Population'],ascending=False).head(7)
g
```

```python
fig, ax = plt.subplots() 
g.plot(y= '2022 Population', x = 'Country', ax = ax) 
g.plot(y = '1970 Population', x = 'Country', ax = ax, secondary_y = True) 
```
